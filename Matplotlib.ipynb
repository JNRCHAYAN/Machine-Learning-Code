# Matplotlib example workflow

```python
%matplotlib inline

import matplotlib.pyplot as plt

# 1. Prepare some data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(5,5))  # Width and height

# 3. Plot data

ax.plot(x,y)

# 4. Customize plot

ax.set(title="Simple Plot",
      xlabel="Width",
      ylabel="Height")

# 5. save and show 
fig.savefig("./image/simple-plot.png")
```

# Two Option for subplots

```python
# Subplot option 1

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,4))

# Plot to each different axis
ax1.plot(x,x/2)
ax1.set(title=("Plot Graph"))
ax2.scatter(np.random.random(10),np.random.random(10))
ax2.set(title=("Scatter Graph"))
ax3.bar(nut_butter_price.keys(),nut_butter_price.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2

fig, ax =plt.subplots(nrows=2,ncols=2,figsize=(10,4))

# Plot to each different Index
ax[0,0].plot(x,x/2)
ax[0,0].set(title=("Plot Graph"))
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[0,1].set(title=("Scatter Graph"))
ax[1,0].bar(nut_butter_price.keys(),nut_butter_price.values())
ax[1,1].hist(np.random.randn(1000));
```

## which one should you use (pyplot vs matplotlib oo method)

* when plotting someting quicjly oke with pyplot
* when plotting something more advance use the oo method

```python
heart_disease.head(5)
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# pyplot Method
over_50.plot(kind="scatter",x ="age",y="chol",c="target")
```

```python
# oo method (Object oriented method)

fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax)
ax.set_xlim([45,80])
```

```python
fig , ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
#customize the plot
ax.set(title="Heart Disease and cholesterol levels",
      xlabel="Age",
      ylabel="Cholesterol")
# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a horisontal line
ax.axhline(over_50["chol"].mean(),
           linestyle= '--')
```

```python
over_50.head()
```

```python
# subplot of chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,10))

# add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
ax0.set(title="Heart Desease levels",
           xlabel="Age",
           ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50["chol"].mean(),
          linestyle="--")

#=========================

# add data to ax1

scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

ax1.set(title="Max Heartbet rate",
           xlabel="Age",
           ylabel="Max Heart Rate")

ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50["thalach"].mean(),
          linestyle="--")

# add a tiitle to the figure

fig.suptitle("Hear Disease Analysis ",fontsize=16,fontweight="bold",horizontalalignment="center")
```
