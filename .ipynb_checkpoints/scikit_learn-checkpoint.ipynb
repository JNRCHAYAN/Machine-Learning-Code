# Introdunction to Sickit_learn (SkLearn)

## An end to end scikit-learn workflow

```python
# 1. Get the data ready

import pandas as pd
heart_disease = pd.read_csv("dataset.csv")
```

```python
heart_disease
```

```python
# Create X (features matrix)
x = heart_disease.drop("target",axis=1)

# Create y(lables)
y = heart_disease["target"]
```

```python
y
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# 2. Choose the right model and hyperparameter

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()

# We will keep the default hyperparamenter
clf.get_params()
```

```python
# 3. Fit the model to training data

from sklearn.model_selection import train_test_split

x_train, x_test,y_train, y_test = train_test_split(x,y,test_size=.2)
```

```python
clf.fit(x_train,y_train)
```

```python
# Make a prediction
import numpy as np

y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
# 4. Evaluate the model

clf.score(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
# 5. improve the model
# Try Different amont of n_estimators

np.random.seed(42)
for i in range(10,100,10):
    print(f"Trying model with {i} esimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train,y_train)
    print(f"Model accouracy on test set :{clf.score(x_test,y_test)*100:.2f}%")
    print(" ")
```

```python
# Save a model

import pickle
```

```python
pickle.dump(clf,open("random_forest_model.pkl","wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model.pkl","rb"))
```

```python
loaded_model.score(x_test,y_test)
```
