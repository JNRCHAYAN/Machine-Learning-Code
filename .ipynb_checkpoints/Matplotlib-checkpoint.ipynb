```python
x = [1,2,3,4,6,8]
y = [11,23,4,5,6,7]
plt.plot(x,y)
```

```python
# 1st Method

fig = plt.figure() # Create a figure
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd Method

fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```
