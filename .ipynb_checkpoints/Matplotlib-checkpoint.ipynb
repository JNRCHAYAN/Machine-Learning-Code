```python
x = [1,2,3,4,6,8]
y = [11,23,4,5,6,7]
plt.plot(x,y)
```

```python
# 1st Method

fig = plt.figure() # Create a figure
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd Method

fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# 3rd Method (Recomanded use)

fig, ax = plt.subplots()
ax.plot(x,y)
type(fig),type(ax)
```

# Matplotlib example workflow

```python
%matplotlib inline

import matplotlib.pyplot as plt

# 1. Prepare some data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(5,5))  # Width and height

# 3. Plot data

ax.plot(x,y)

# 4. Customize plot

ax.set(title="Simple Plot",
      xlabel="Width",
      ylabel="Height")

# 5. save and show 
fig.savefig("./image/simple-plot.png")
```

## Making Figures with numpy arrays

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0,10,100)
x
```

```python
#plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x,x**2);
```

```python
# Use the same data scatter plot

fig , ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
# Another Scatter plot

fig,ax = plt.subplots()

ax.scatter(x, np.sin(x))
```

```python
# Make a plot from dictornary

nut_butter_price = {
            "Almond Butter" : 20,
            "Peanut Butter" : 8,
            "Cashew Butter" : 12,
            "Kaxi Butter" :  24,
}

fig , ax = plt.subplots(figsize=(7,7))
ax.bar(nut_butter_price.keys(),nut_butter_price.values())
ax.set(

    title="Butter Store",
    xlabel="Butter Name",
    ylabel="Price",
)
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_price.keys(),nut_butter_price.values())
ax.set(
    title="Butter Store",
    xlabel="Price",
    ylabel="Butter Name")
```

```python
# Make Some data and histagram

x = np.random.randn(1000)
fig , ax = plt.subplots()

ax.hist(x)
```

# Two Option for subplots

```python
# Subplot option 1

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,4))

# Plot to each different axis
ax1.plot(x,x/2)
ax1.set(title=("Plot Graph"))
ax2.scatter(np.random.random(10),np.random.random(10))
ax2.set(title=("Scatter Graph"))
ax3.bar(nut_butter_price.keys(),nut_butter_price.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2

fig, ax =plt.subplots(nrows=2,ncols=2,figsize=(10,4))

# Plot to each different Index
ax[0,0].plot(x,x/2)
ax[0,0].set(title=("Plot Graph"))
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[0,1].set(title=("Scatter Graph"))
ax[1,0].bar(nut_butter_price.keys(),nut_butter_price.values())
ax[1,1].hist(np.random.randn(1000));
```
